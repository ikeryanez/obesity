```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../data/obesity_red.csv')
```

```python
# Graficar las variables numÃ©ricas
numeric_vars = df.select_dtypes(include=['int64', 'float64'])
sns.pairplot(numeric_vars)
plt.show()

# Graficar las variables categÃ³ricas
categorical_vars = df.select_dtypes(include=['object', 'category'])
for column in categorical_vars.columns:
    sns.countplot(x=column, data=df)
    plt.show()
```
