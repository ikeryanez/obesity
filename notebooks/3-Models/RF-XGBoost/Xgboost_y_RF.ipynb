# CODIFICAR VARIABLE OBJETIVO

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
nobeyesdad_categories = ['Insufficient_Weight', 'Normal_Weight', 'Overweight_Level_I', 'Overweight_Level_II', 'Obesity_Type_I', 'Obesity_Type_II', 'Obesity_Type_III']

# Inicializa el codificador ordinal
ordinal_encoder = OrdinalEncoder(categories=[nobeyesdad_categories])

# Aplica el codificador ordinal a las columnas específicas
train[['nobeyesdad']] = ordinal_encoder.fit_transform(train[['nobeyesdad']])

# Muestra las primeras filas del DataFrame para verificar los cambios
print(train.head())
```

```python
nobeyesdad_categories = ['Insufficient_Weight', 'Normal_Weight', 'Overweight_Level_I', 'Overweight_Level_II', 'Obesity_Type_I', 'Obesity_Type_II', 'Obesity_Type_III']

# Inicializa el codificador ordinal
ordinal_encoder = OrdinalEncoder(categories=[nobeyesdad_categories])

# Aplica el codificador ordinal a las columnas específicas
test[['nobeyesdad']] = ordinal_encoder.fit_transform(test[['nobeyesdad']])

# Muestra las primeras filas del DataFrame para verificar los cambios
print(test.head())
```

# FEATURE ENGINEERING

```python
train['height_m'] = train['height'] / 100 #pasar_height a cm

# Ahora calcula el BMI:
train['bmi'] = train['weight'] / (train['height_m']**2)
```

```python
test['height_m'] = test['height'] / 100 #pasar_height a cm

# Ahora calcula el BMI:
test['bmi'] = test['weight'] / (test['height_m']**2)
```

# MODELO XGBOOST

```python
pip install xgboost scikit-learn
```

```python
X_train = train.drop('nobeyesdad', axis=1)
y_train = train['nobeyesdad']
```

```python
X_test = test.drop('nobeyesdad', axis=1)
y_test = test['nobeyesdad']
```

```python
import xgboost as xgb

# Configurar el clasificador
clf = xgb.XGBClassifier(
    objective='multi:softprob',  # Usado para clasificación multiclase
    eval_metric='mlogloss',      # Métrica de evaluación para clasificación multiclase
    num_class=len(y_train.unique()),   # Número de clases
    learning_rate=0.1,           # Tasa de aprendizaje
    n_estimators=100,            # Número de árboles
    max_depth=6,                 # Profundidad máxima de cada árbol
    subsample=0.8,               # Submuestra de la población de entrenamiento
    colsample_bytree=0.8         # Submuestra de características
)

# Entrenar el modelo
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Predecir las etiquetas para el conjunto de prueba
y_pred = clf.predict(X_test)

# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Mostrar un reporte de clasificación detallado
print(classification_report(y_test, y_pred))
```

Sin el feature enginnering: Accuracy: 97.66%
