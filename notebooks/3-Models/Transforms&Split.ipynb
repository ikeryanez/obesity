```python
gender_dict = {'Female':0, 'Male':1}
df['gender'] = df['gender'].map(gender_dict)

hist_dict = {'no':0, 'yes':1}
df['family_history_with_overweight'] = df['family_history_with_overweight'].map(hist_dict)

favc_dict = {'no':0, 'yes':1}
df['favc'] = df['favc'].map(favc_dict)

caec_dict = {'no':0, 'Sometimes':1, 'Frequently':2, 'Always':3}
df['caec'] = df['caec'].map(caec_dict)

smoke_dict = {'no':0, 'yes':1}
df['smoke'] = df['smoke'].map(smoke_dict)

scc_dict = {'no':0, 'yes':1}
df['scc'] = df['scc'].map(scc_dict)

calc_dict = {'no':0, 'Sometimes':1, 'Frequently':2, 'Always':3}
df['calc'] = df['calc'].map(calc_dict)

trans_dict = {'Automobile':0, 'Motorbike':1, 'Public_Transportation':2, 'Bike':3, 'Walking':4}
df['mtrans'] = df['mtrans'].map(trans_dict)

# obesity_dict = {'Insufficient_Weight':0, 'Normal_Weight':1, 'Overweight_Level_I':2, 'Overweight_Level_II':3, 'Obesity_Type_I':4, 'Obesity_Type_II':5, 'Obesity_Type_III':6}
# df['nobeyesdad'] = df['nobeyesdad'].map(obesity_dict)
```

```python
from sklearn.model_selection import train_test_split

# Dividir los datos en conjuntos de entrenamiento y prueba
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
train_df.shape, test_df.shape
```

```python
train_df.to_csv('train.csv', index=False)
test_df.to_csv('test.csv', index=False)
```

```python
train_regression_df = train_df.copy(deep=True)
test_regression_df = test_df.copy(deep=True)
```

```python
obesity_dict = {'Insufficient_Weight':0, 'Normal_Weight':1, 'Overweight_Level_I':2, 'Overweight_Level_II':3, 'Obesity_Type_I':4, 'Obesity_Type_II':5, 'Obesity_Type_III':6}
```

```python
train_regression_df['nobeyesdad'] = train_regression_df['nobeyesdad'].map(obesity_dict)
test_regression_df['nobeyesdad'] = test_regression_df['nobeyesdad'].map(obesity_dict)
```

```python
train_regression_df.to_csv('train_regression.csv', index=False)
test_regression_df.to_csv('test_regression.csv', index=False)
```
